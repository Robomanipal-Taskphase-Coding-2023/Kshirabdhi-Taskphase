```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file ="F:/xyz/dataset.csv"
df = pd.read_csv(file)
print(df)
```

```python


# Assuming 'df' is your DataFrame and 'categorical_columns' is a list of columns you want to one-hot encode
categorical_columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm']

# One-hot encoding categorical columns
df = pd.get_dummies(df, columns=categorical_columns)

# Displaying the result
print(df)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'binary_columns' is a list of columns with "yes" or "no"
binary_columns = ['RainToday', 'RainTomorrow']

# One-hot encoding binary columns without dropping the first category
df = pd.get_dummies(df, columns=binary_columns, drop_first=False)

# Displaying the result
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame and 'Location' is the column to be encoded
le = LabelEncoder()
df['Location'] = le.fit_transform(df['Location'])


```

```python

print(df)
```

```python
x= df.iloc[:,1 :-1]
```

```python
y = df.iloc[:, -1]
```

```python
y
```

```python
x
```

```python
### train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
     x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Preprunning
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_test
```

```python
y_pred=cv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
